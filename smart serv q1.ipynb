```python
response = requests.get("https://s3.amazonaws.com/open-to-cors/assignment.json")
todos = json.loads(response.text)
```

```python
todos
```

```python
y=todos['products']
    
```

```python
print(y)
```

```python
for i in y:
    z=y[i]
    z['popularity']=int(z['popularity'])
```

```python
y
```

```python
final=sorted(y.items(), key=lambda x: x[1]['popularity'], reverse=True)
final=dict(final)
```

```python
final
```

```python
with open('smart.csv', 'w') as csvfile:
        for key in final.keys():
            csvfile.write("%s,%s\n"%(key,final[key]))
```
